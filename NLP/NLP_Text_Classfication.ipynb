```python
from sklearn.datasets import fetch_20newsgroups
```

```python
print(dict(fetch_20newsgroups()).keys()) #어떤 정보를 가지고 있는가
print(len(fetch_20newsgroups().data)) #길이
print(fetch_20newsgroups().target_names[0])
```

```python
for cate in fetch_20newsgroups().target_names:
  print(f'category {cate}')
```

```python
# train data
categories = ['alt.atheism', 'comp.graphics', 'sci.med', 'soc.religion.christian']

twenty_train = fetch_20newsgroups(
    subset='train',
    categories = categories,
    shuffle= True)
```

```python
print(f'keys: {dict(twenty_train).keys()}')
print(f'length: {len(twenty_train.data)}')
print(twenty_train.target_names[0])
```

```python
# Extracting Features from Text
# Vectorization ex. CountVector, TF-IDF etc

from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer()
X_train_counts = count_vector.fit_transform(twenty_train.data)
```

```python
print(X_train_counts.shape) #(train data수, 단어들의 모음)
```

```python
print(X_train_counts[0].shape)
```

```python
print(X_train_counts[0]) # (0번째 데이터, 피쳐번호) 나온 횟수(카운트)
```

```python
print(count_vector.vocabulary_) # 5번 등장한 32142는 The임
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfTransformer

tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_counts)
X_train_tf = tf_transformer.transform(X_train_counts)
```

```python
print(X_train_tf.shape)
print(X_train_tf[0])
```

```python
# Training a Classifier - Naive Baysian
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train_tf, twenty_train.target) # model생성
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
pipe_clf = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', MultinomialNB()),
])
pipe_clf.fit(twenty_train.data, twenty_train.target)
```

```python
#Evaluation
# test dataset 생성
twenty_test = fetch_20newsgroups(
    subset = 'test',
    categories = categories,
    shuffle = True
)
```

```python
import numpy as np

predicted = pipe_clf.predict(twenty_test.data)
result = np.mean(predicted == twenty_test.target)
print(f'Accuracy: {result*100}')
```

```python
from collections import Counter
print(Counter(twenty_train.target))
print(Counter(twenty_test.target))   #분포가 다를 경우 F1 스코어 확인
```

```python
from sklearn import metrics

result = metrics.classification_report(
    y_true = twenty_test.target,
    y_pred = predicted,
    target_names = twenty_test.target_names)

print(result)
```
